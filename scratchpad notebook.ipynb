Import matplotlib

```python
import matplotlib.pyplot as plt
```

Set the Coordinates to use

```python
x_coords = [0, 1, 2, 3, 4]
y_coords = [0, 3, 1, 5, 2] 
```

Plot first graph

```python
# plot the graph with markers
plt.plot(x_coords, y_coords, marker='o')

# add title
plt.title('Sample Data')

# Add Labels to the axis
plt.xlabel('X axis')
plt.ylabel('Y axis')

# change limits of the graph
plt.xlim(xmin=-1, xmax=10)
plt.ylim(ymin=-1, ymax=6)

# add tic marks
plt.xticks([0, 1, 2, 3, 4],
           ['2016', '2017', '2018', '2019', '2020'])

# show grid
plt.grid(True)

# show the graph
plt.show()
```

Plotting Bar Charts

```python
bar_width = .4

plt.bar(x_coords, y_coords, bar_width, color=('b', 'g', 'b', 'r', 'k'))

plt.show()
```

Pie Chart

```python
values = [20, 60, 80, 40]
my_labels = ['Q1', 'Q2', 'Q3', 'Q4']
plt.pie(values, labels=my_labels)
plt.show()
```
